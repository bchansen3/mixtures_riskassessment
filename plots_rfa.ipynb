## This file reproduces some the plots included in the *_Utilizing Human Exposure and Biomonitoring Data to Improve the Application of New Approach Methods in Fit-For-Purpose Mixture Assessment_* proposal.

### To make a bimodal distribution

```r
library(tidyverse)
library(ggplot2)
set.seed(4)

par(mfrow=c(2,1))

bimod <- c(rnorm(1000, 0, 1), rnorm(500, -4, 1))
  ggplot(mapping = aes(bimod)) +geom_density(color ="blue",fill ="lightblue")+
    geom_vline(aes(xintercept=c(-4,0)), color="blue",
             linetype="dashed")+
      theme_classic()+
      theme(axis.title.x=element_blank(),
        axis.text.x=element_blank(),
        axis.ticks.x=element_blank(),
        axis.title.y=element_blank(),
        axis.text.y=element_blank(),
        axis.ticks.y=element_blank())

onemod <- c(rnorm(1500, 0, 1))
    ggplot(mapping = aes(onemod)) +geom_density(color ="blue",fill ="lightblue")+
      theme_classic()+
      theme(axis.title.x=element_blank(),
        axis.text.x=element_blank(),
        axis.ticks.x=element_blank(),
        axis.title.y=element_blank(),
        axis.text.y=element_blank(),
        axis.ticks.y=element_blank())

```

```r
par(mfrow=c(2,1))

set.seed(4)

samp_bimod <- sample(bimod, size=500, replace=T)

samp_onemod <- sample(onemod, size=500, replace=T)


  ggplot(mapping = aes(samp_bimod)) +geom_histogram(color ="blue",fill ="lightblue")+
      theme_classic()+
      theme(axis.title.x=element_blank(),
        axis.text.x=element_blank(),
        axis.ticks.x=element_blank(),
        axis.title.y=element_blank(),
        axis.text.y=element_blank(),
        axis.ticks.y=element_blank()) 

onemod <- c(rnorm(1500, 0, 1))
  ggplot(mapping = aes(samp_onemod)) +geom_histogram(color ="blue",fill ="lightblue")+
      theme_classic()+
      theme(axis.title.x=element_blank(),
        axis.text.x=element_blank(),
        axis.ticks.x=element_blank(),
        axis.title.y=element_blank(),
        axis.text.y=element_blank(),
        axis.ticks.y=element_blank())
```
